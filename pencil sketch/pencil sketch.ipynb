```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
```

```python
def dodge(front,back):
    result=front*255/(255-back) 
    result[result>255]=255
    result[back==255]=255
    return result.astype('uint8')
```

```python
def grayscale(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
img='viratkohli.jpg'
s = imageio.imread(img)
g=grayscale(s)
i = 255-g
import scipy.ndimage
b = scipy.ndimage.filters.gaussian_filter(i,sigma=8)
r= dodge(b,g)
```

```python
%matplotlib inline 
plt.imshow(r, cmap="gray")
```

```python
plt.imsave('pencil.png', r, cmap='gray', vmin=0, vmax=255)
```

```python
plt.imsave('greyscale.png', g, cmap='gray', vmin=0, vmax=255)
plt.imshow(g, cmap="gray")
```

```python
plt.imsave('green_blue.png', r,cmap='winter')
plt.imshow( r,cmap='winter')
```
